### Ex1:Create a figure object called fig using plt.figure() , Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)  
y = np.sin(x)  

fig, ax = plt.subplots()

ax.plot(x, y, label='Sin(x)', color='green')

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Plot of Sin(x)')

plt.show()
```

### Ex4:Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

def draw_line_from_file(file_path, x_label, y_label, title):
    with open('data.txt', 'r') as file:
        lines = file.readlines()

    x_values = []
    y_values = []
    for line in lines:
        x, y = map(float, line.strip().split())
        x_values.append(x)
        y_values.append(y)

    
    plt.plot(x_values, y_values)

   
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)

    
    plt.show()


file_path = 'data.txt'  
x_label = 'X-axis'
y_label = 'Y-axis'
title = 'Line Plot'

draw_line_from_file(file_path, x_label, y_label, title)
```

### Ex5:Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt

def plot_lines(x_values, y_values_list, labels, colors, widths):
    for y_values, label, color, width in zip(y_values_list, labels, colors, widths):
        plt.plot(x_values, y_values, label=label, color=color, linewidth=width)

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Multiple Lines Plot')
    plt.legend()
    plt.show()

x_values = [1, 2, 3, 4, 5]
y_values_list = [ [2, 4, 6, 8, 10],[1, 3, 5, 7, 9],[0, 2, 4, 6, 8]]
labels = ['Line 1', 'Line 2', 'Line 3']
colors = ['blue', 'green', 'red']
widths = [2, 1.5, 2.5]

plot_lines(x_values, y_values_list, labels, colors, widths)
```

### Ex6:Write a Python program to plot quantities which have an x and y position.

```python
import matplotlib.pyplot as plt

def plot_quantities(x_values, y_values):
    plt.scatter(x_values, y_values, marker='o', s=50)

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Scatter Plot of Quantities')
    plt.show()
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 3, 5]

plot_quantities(x_values, y_values)
```

### Ex7: Create plot usingsubplots(),scatter, and some other methods on theAxesobject

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x_values = np.random.rand(50)
y_values = np.random.rand(50)

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

#Scatter plot
axes[0, 0].scatter(x_values, y_values, marker='o', color='blue')
axes[0, 0].set_title('Scatter Plot')

#Line plot
axes[0, 1].plot(x_values, y_values, color='green', linestyle='--')
axes[0, 1].set_title('Line Plot')

#Bar plot
axes[1, 0].bar(range(len(x_values)), y_values, color='orange')
axes[1, 0].set_title('Bar Plot')

#Histogram
axes[1, 1].hist(x_values, bins=20, color='purple', alpha=0.7)
axes[1, 1].set_title('Histogram')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Ex10: Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = np.random.randn(1000)  

plt.hist(data, bins=20, color='green', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Dataset')

plt.show()
```

### Ex11:Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.

```python
import matplotlib.pyplot as plt

expense_categories = ['Rent', 'Utilities', 'Groceries', 'Entertainment', 'Transportation']
expenses = [1500, 300, 500, 200, 400]

plt.pie(expenses, labels=expense_categories, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink'])

plt.title('Expense Distribution in Budget')

plt.show()
```
